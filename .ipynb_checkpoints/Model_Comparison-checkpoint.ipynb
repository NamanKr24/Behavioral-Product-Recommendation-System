```python
import pandas as pd
from collections import defaultdict
import warnings
warnings.filterwarnings('ignore')
from data_loader import load_cosmetics_data
from data_preprocessing import preprocess_data
from evaluation import precision_at_k, recall_at_k
from baseline_model import get_baseline_recommendations
from content_model import train_content_based_model, get_content_recommendations
from collaborative_model import train_collaborative_model, get_collaborative_recommendations
import matplotlib.pyplot as plt
import seaborn as snsS
```

```python
raw_df = load_cosmetics_data()
raw_df['event_time'] = pd.to_datetime(raw_df['event_time'])
df = preprocess_data(raw_df)
print("Data ready.")
```

```python
df_sorted = df.sort_values(by=['user_id', 'event_time'])
purchases_df = df_sorted[df_sorted['event_type'] == 'purchase']

test_df = purchases_df.groupby('user_id').tail(1)
train_df = df.drop(test_df.index)

actuals = test_df.groupby('user_id')['product_id'].apply(list).to_dict()
print(f"Split complete. Test set contains {len(actuals)} users.")
```

```python
baseline_recs = get_baseline_recommendations(train_df, n=10)
cb_nn_model, cb_tfidf_matrix, cb_product_metadata, cb_indices = train_content_based_model(train_df)

cf_algo = train_collaborative_model(train_df)
print("All models trained.")
```

```python
K = 10
results = defaultdict(lambda: {'precision': [], 'recall': []})

all_product_ids = train_df['product_id'].unique()
```

```python
for user_id, actual_items in actuals.items():
    results['Baseline']['precision'].append(precision_at_k(K, baseline_recs, actual_items))
    results['Baseline']['recall'].append(recall_at_k(K, baseline_recs, actual_items))
    
    user_train_history = train_df[train_df['user_id'] == user_id]
    last_interacted_product = user_train_history['product_id'].iloc[-1]
    content_recs = get_content_recommendations(last_interacted_product, cb_nn_model, cb_tfidf_matrix, cb_product_metadata, cb_indices, n=K)
    results['Content-Based']['precision'].append(precision_at_k(K, content_recs, actual_items))
    results['Content-Based']['recall'].append(recall_at_k(K, content_recs, actual_items))

    interacted_items = train_df[train_df['user_id'] == user_id]['product_id'].unique()
    collab_recs = get_collaborative_recommendations(user_id, all_product_ids, interacted_items, cf_algo, n=K)
    results['Collaborative']['precision'].append(precision_at_k(K, collab_recs, actual_items))
    results['Collaborative']['recall'].append(recall_at_k(K, collab_recs, actual_items))
```

```python
results_list = []

for model_name, scores in results.items():
    avg_precision = sum(scores['precision']) / len(scores['precision'])
    avg_recall = sum(scores['recall']) / len(scores['recall'])
    results_list.append({'Model': model_name, 
                         'Avg_Precision@10': avg_precision, 
                         'Avg_Recall@10': avg_recall})

summary_df = pd.DataFrame(results_list)

print(summary_df.to_string(index=False))
```

```python
df_long = pd.melt(summary_df, id_vars=['Model'], var_name='Metric', value_name='Score')

plt.figure(figsize=(12, 7))
ax = sns.barplot(x='Model', y='Score', hue='Metric', data=df_long, palette='viridis')

plt.title('Model Performance Comparison', fontsize=16)
plt.xlabel('Recommendation Model', fontsize=12)
plt.ylabel('Average Score', fontsize=12)
plt.ylim(0, max(df_long['Score']) * 1.2)
plt.legend(title='Metric', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
